```python
import re
from konlpy.tag import Kkma

kkma = Kkma()


def jaccard_similarity(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union

#threshold = 5 # Define a threshold for similarity
threshold = 0.3 # Define a threshold for similarity

lines_seen = set() # Define an empty set to keep track of lines

in_file = r'/content/drive/MyDrive/datasets/spam-message2.csv'
#out_file = r'/content/drive/MyDrive/datasets/xspam-message2.csv'
out_file = r'/content/xspam-message2.csv'

i = 0
my_names = ['손학규', '정동영','김쌍우','박양주', '강성만', '이종담', '강명구'] 
print("\n\n\Before \n")
with open(in_file, 'r') as f_in, open(out_file, 'w') as f_out:
    for line in f_in:
        linex = line
        line = line.strip().lower() # Strip off whitespace and convert to lowercase
        names = my_names# Names names names!!!
        for name in names:
            if name in line : # Check if the entity is a person name
                line = re.sub(re.escape(name), 'ooo', line) # Remove each name from the text

        line = re.sub(r'\d+', '', line) # Remove numerical characters from the line
        line_set = set(line.split())
        is_similar = False
        print(i, line)

        j = 0
        for seen_line in lines_seen:
            seen_line = re.sub(r'\d+', '', seen_line) # Remove numerical characters from the seen line
            seen_names = my_names # Extract the names from the seen line
            for name in seen_names:
                if name in seen_line : # Check if the entity is a person name
                    seen_line = re.sub(re.escape(name), 'ooo', seen_line) # Remove each name from the seen line
            seen_line_set = set(seen_line.split())
            similarity = jaccard_similarity(line_set, seen_line_set)
            if similarity > threshold:
                is_similar = True
                #print(" Similarity test at ", j, " , Filtered?  ", is_similar, '\n')
                break
            j += 1
            

        if(i>50) :
          break
        else :
          i += 1

        if not is_similar:
            lines_seen.add(line)
            #print("Lines add, ", lines_seen)
            f_out.write(linex + '\n')
            #print("Output lines, ", linex)

i=0
print("\n\n\nAfter\n ")
for line in lines_seen:
  print(i, line)
  i += 1
```
